```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def softmax_temperature(x, temperature=1.):
    return -np.exp(x/temperature)/ (-np.sum(np.exp(x/temperature)))
```

```python
result = np.array([2.4, 3.2, 5.6, 4.7, 0.9, 2.1, 1.6])
ind = np.arange(7)
fig=plt.figure(num=1, figsize=(8, 8))
ax1 = fig.add_subplot(331)
ax2 = fig.add_subplot(332)
ax3 = fig.add_subplot(333)
ax5 = fig.add_subplot(335)

ax7 = fig.add_subplot(337)
ax8 = fig.add_subplot(338)
ax9 = fig.add_subplot(339)

ax1.set_ylim([0,1])
ax2.set_ylim([0,1])
ax3.set_ylim([0,1])

ax5.set_ylim([0,1])

ax7.set_ylim([0,1])
ax8.set_ylim([0,1])
ax9.set_ylim([0,1])

ax1.bar(ind,softmax_temperature(result, 0.1),0.5)
ax2.bar(ind,softmax_temperature(result, 0.4),0.5)
ax3.bar(ind,softmax_temperature(result, 0.7),0.5)

ax5.bar(ind,softmax_temperature(result, 1),0.5)
ax7.bar(ind,softmax_temperature(result, 1.2),0.5)
ax8.bar(ind,softmax_temperature(result, 1.4),0.5)
ax9.bar(ind,softmax_temperature(result, 1.6),0.5)
plt.show()
```
