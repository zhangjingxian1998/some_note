```python
# 定义函数为y=\alpha x, \alpha = 2
# 初始化\alpha=0
# 先不使用学习率
# 使用loss (y-y_)^2
# 采样一个数据点(1,2)

x= 10000
y = 2 * x
alpha = 0

def gd(x, y, y_):
    alpha = -2 * x * (y - y_)
    return alpha

def gd_lr(x, y, y_, lr=0.1):
    alpha = -2 * x * (y - y_) * lr
    return alpha

# for i in range(1000):
#     y_ = alpha * x
#     alpha_delta = gd(x, y, y_)
#     alpha -= alpha_delta
#     print(f'epoch: {i+1}, alpha: {alpha}')

for i in range(1000):
    y_ = alpha * x
    alpha_delta = gd_lr(x, y, y_)
    alpha -= alpha_delta
    print(f'epoch: {i+1}, alpha: {alpha}')
```
