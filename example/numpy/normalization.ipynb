```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# 生成随机数据
dim_one_random_data = np.array([np.random.random() for i in range(10)])

dim_two_random_data_x = np.array([np.random.random() for i in range(50)])
dim_two_random_data_y = np.array([np.random.random() for i in range(50)])
dim_two_random_data_y = dim_two_random_data_y * 6
```

```python
plt.scatter(dim_one_random_data, np.ones(len(dim_one_random_data)))
mean_dim_one_random_data = np.mean(dim_one_random_data)
dim_one_random_data_mean = (dim_one_random_data - mean_dim_one_random_data)
std_dim_one_random_data = np.std(dim_one_random_data_mean)
dim_one_random_data_norm = dim_one_random_data_mean / std_dim_one_random_data
plt.scatter(dim_one_random_data_norm, np.ones(len(dim_one_random_data))*1.2)
```

```python
plt.xlim([-7,7])
plt.ylim([-7,7])
plt.scatter(dim_two_random_data_x, dim_two_random_data_y)
mean_x = np.mean(dim_two_random_data_x)
mean_y = np.mean(dim_two_random_data_y)
dim_two_random_data_x_mean = dim_two_random_data_x - mean_x
dim_two_random_data_y_mean = dim_two_random_data_y - mean_y

std_x = np.std(dim_two_random_data_x_mean)
std_y = np.std(dim_two_random_data_y_mean)

dim_two_random_data_x_norm = dim_two_random_data_x_mean / std_x
dim_two_random_data_y_norm = dim_two_random_data_y_mean / std_y
plt.scatter(dim_two_random_data_x_norm, dim_two_random_data_y_norm)
```

```python
w1 = 6
b1 = 0
w2 = 1
b2 = 0

random_data_x = np.array([np.random.random() for _ in range(100)])
random_data_y = np.array([np.random.random() for _ in range(100)]) * 6
random_data_z = w1*random_data_x + w2*random_data_y
fig=plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(random_data_x, random_data_y, random_data_z, cmap='spectral', c='r', s=50)
save_data = np.stack([random_data_x,random_data_y,random_data_z],axis=1)
np.save('../data/z=x+y.npy',save_data)
```
